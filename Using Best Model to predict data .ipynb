```python
import joblib
import pandas as pd
```

```python
file="Random Forest Regressor.sav"
```

```python
sav_model=joblib.load(file)
```

```python
testdataset="C:/Users/abhis/OneDrive/Desktop/ML Test/test.csv"
data_test=pd.read_csv(testdataset)
data_test.head(100)
```

```python
data_test.rename(columns={'count':'total_count'},inplace=True)
data_test['datetime']=pd.to_datetime(data_test.datetime)
data_test['season']=data_test.season.astype('category')
data_test['holiday']=data_test.holiday.astype('category')
data_test['workingday']=data_test.workingday.astype('category')
data_test['weather']=data_test.weather.astype('category')
```

```python
X=data_test[["temp","humidity","windspeed","season","holiday","workingday","weather"]]
cat_attributes=['season','holiday','workingday','weather']
num_attributes=['temp','humidity','windspeed']
```

```python
test_encoded_attributes=pd.get_dummies(X,columns=cat_attributes)
print('Shape of transfomed dataframe::',test_encoded_attributes.shape)
test_encoded_attributes.head(5)
```

```python
X_test=test_encoded_attributes
X_test.head()
```

```python
pred=sav_model.predict(X_test)
```

```python
data_test['Predicted count']=pred
```
